## Checking NUll values

```python
RawData.isnull().sum()
```

```python
# Now filling the numeric type columns with mean or median
RawData['result_margin']=RawData['result_margin'].fillna(RawData['result_margin'].median())
RawData['target_runs']=RawData['target_runs'].fillna(RawData['target_runs'].mean())
RawData['target_overs']=RawData['target_overs'].fillna(RawData['target_overs'].mean())

# Now filling Object type columns with mode
RawData['city']=RawData['city'].fillna(RawData['city'].mode()[0])
RawData['player_of_match']=RawData['player_of_match'].fillna(RawData['player_of_match'].mode()[0])
RawData['winner']=RawData['winner'].fillna(RawData['winner'].mode()[0]) 
RawData['method'] = RawData['method'].fillna('Normal')
```

```python
RawData.isnull().sum()
```

## Checking for duplicates values

```python
RawData.duplicated().sum()

if RawData.duplicated().sum()==0:
    print("No duplicate values found in the dataset.")
else:
    print("Duplicate values found in the dataset. Removing duplicates...")
    RawData.drop_duplicates(inplace=True)
```

## Converting 

```python
RawData['date'] = pd.to_datetime(RawData['date'], errors='coerce')
RawData['season'] = RawData['date'].dt.year
```

```python
RawData.info()
```

## chaning Data type of Super_over column

```python
RawData['super_over'] = RawData['super_over'].map({
    'Y': 1,
    'N': 0
}).astype(int)
```

## Match Insights

### **Matches Played Per Season**

```python
# For Final
final_matches = RawData[RawData['match_type'] == 'Final']
final_teams = pd.concat([final_matches['team1'], final_matches['team2']])
final_team_counts = final_teams.value_counts()
print("Teams with most Final appearances:")
print(final_team_counts.head(5))

# For Qualifier 1
qual1_matches = RawData[RawData['match_type'] == 'Qualifier 1']
qual1_teams = pd.concat([qual1_matches['team1'], qual1_matches['team2']])
qual1_team_counts = qual1_teams.value_counts()
print("\nTeams with most Qualifier 1 appearances:")
print(qual1_team_counts.head(5))

# For Qualifier 2
qual2_matches = RawData[RawData['match_type'] == 'Qualifier 2']
qual2_teams = pd.concat([qual2_matches['team1'], qual2_matches['team2']])
qual2_team_counts = qual2_teams.value_counts()
print("\nTeams with most Qualifier 2 appearances:")
print(qual2_team_counts.head(5))
```

### **Biggest Win Margins**

```python
top_margins = RawData['result_margin'].value_counts().head(10).index

for margin in top_margins:

    teams_for_margin = RawData[RawData['result_margin'] == margin]['winner'].value_counts().head(5)

    print(f"\nMargin {margin}: {teams_for_margin.to_dict()}")
```
